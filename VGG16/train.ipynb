```python
import os
import rarfile
from PIL import Image
import numpy as np
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# Lấy đường dẫn hiện tại của notebook
current_dir = os.getcwd()

# Đường dẫn tới tệp RAR train_normal.rar
train_normal_rar_file = os.path.join(current_dir, 'train_normal.rar')

# Giải nén tệp RAR train_normal.rar
extract_dir_train_normal = os.path.join(current_dir, 'train_normal')
with rarfile.RarFile(train_normal_rar_file, 'r') as rf:
    rf.extractall(extract_dir_train_normal)

# Chuẩn bị danh sách các tệp ảnh trong thư mục giải nén
train_normal_image_files = []
for file_name in os.listdir(extract_dir_train_normal):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png'):
        train_normal_image_files.append(os.path.join(extract_dir_train_normal, file_name))

# Tiền xử lý dữ liệu huấn luyện
x_train = []
y_train = []
for image_file in train_normal_image_files:
    # Đọc tệp ảnh
    image = Image.open(image_file)
    # Chuyển đổi thành mảng numpy
    image = np.array(image)
    # Gán ảnh vào mảng x_train
    x_train.append(image)
    # Gán nhãn vào mảng y_train (ví dụ: 0 trong trường hợp này)
    y_train.append(0)
```

```python
# Lấy đường dẫn hiện tại của notebook
current_dir = os.getcwd()

# Đường dẫn tới tệp RAR train_abnormal.rar
train_abnormal_rar_file = os.path.join(current_dir, 'train_abnormal.rar')

# Giải nén tệp RAR train_abnormal.rar
extract_dir_train_abnormal = os.path.join(current_dir, 'train_abnormal')
with rarfile.RarFile(train_abnormal_rar_file, 'r') as rf:
    rf.extractall(extract_dir_train_abnormal)

# Chuẩn bị danh sách các tệp ảnh trong thư mục giải nén
train_abnormal_image_files = []
for file_name in os.listdir(extract_dir_train_abnormal):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png'):
        train_abnormal_image_files.append(os.path.join(extract_dir_train_abnormal, file_name))

for image_file in train_abnormal_image_files:
    # Đọc tệp ảnh
    image = Image.open(image_file)
    # Chuyển đổi thành mảng numpy
    image = np.array(image)
    # Gán ảnh vào mảng x_train
    x_train.append(image)
    # Gán nhãn vào mảng y_train (ví dụ: 1 trong trường hợp này)
    y_train.append(1)
```

```python
import shutil

# Đường dẫn thư mục lưu trữ ảnh trên máy tính
download_dir = os.path.join(current_dir, 'downloaded_images')

# Tạo thư mục lưu trữ ảnh nếu chưa tồn tại
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Download các ảnh từ x_train
for i, image in enumerate(x_train):
    # Tạo đường dẫn tới tệp ảnh
    image_path = os.path.join(download_dir, f'image_{i}.jpg')

    # Lưu ảnh với đường dẫn đã tạo
    Image.fromarray(image).save(image_path)

print(f'Các ảnh đã được tải xuống thành công và lưu tại: {download_dir}')
```

```python
# Chuẩn bị dữ liệu huấn luyện và nhãn tương ứng
x_train = np.array(x_train)
y_train = np.array(y_train)
x_train = x_train / 255.0

# Xây dựng mô hình Sequential trên mô hình VGG-16
model = Sequential()
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(224, 224, 3)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Tạo đối tượng tối ưu hóa SGD và thiết lập learning rate
# sgd = SGD(learning_rate=1e-4, momentum=0.9)

# Compile và huấn luyện mạng
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=25, batch_size=64)
```

```python
# Đường dẫn tới tệp RAR test_normal.rar
test_normal_rar_file = os.path.join(current_dir, 'test_normal.rar')

# Giải nén tệp RAR test_normal.rar
extract_dir_test_normal = os.path.join(current_dir, 'test_normal')
with rarfile.RarFile(test_normal_rar_file, 'r') as rf:
    rf.extractall(extract_dir_test_normal)

# Chuẩn bị danh sách các tệp ảnh trong thư mục giải nén
test_normal_image_files = []
for file_name in os.listdir(extract_dir_test_normal):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png'):
        test_normal_image_files.append(os.path.join(extract_dir_test_normal, file_name))

# Tiền xử lý dữ liệu kiểm tra
x_test = []
for image_file in test_normal_image_files:
    # Đọc tệp ảnh
    image = Image.open(image_file)
    # Chuyển đổi thành mảng numpy
    image = np.array(image)
    # Gán ảnh vào mảng x_test
    x_test.append(image)
```

```python
# Đường dẫn tới tệp RAR test_abnormal.rar
test_abnormal_rar_file = os.path.join(current_dir, 'test_abnormal.rar')

# Giải nén tệp RAR test_abnormal.rar
extract_dir_test_abnormal = os.path.join(current_dir, 'test_abnormal')
with rarfile.RarFile(test_abnormal_rar_file, 'r') as rf:
    rf.extractall(extract_dir_test_abnormal)

# Chuẩn bị danh sách các tệp ảnh trong thư mục giải nén
test_abnormal_image_files = []
for file_name in os.listdir(extract_dir_test_abnormal):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png'):
        test_abnormal_image_files.append(os.path.join(extract_dir_test_abnormal, file_name))

for image_file in test_abnormal_image_files:
    # Đọc tệp ảnh
    image = Image.open(image_file)
    # Chuyển đổi thành mảng numpy
    image = np.array(image)
    # Gán ảnh vào mảng x_test
    x_test.append(image)
```

```python
# Chuẩn bị dữ liệu kiểm tra
x_test = np.array(x_test)
x_test = x_test / 255.0

# Dự đoán nhãn
y_test = model.predict(x_test)

test_image_files = test_normal_image_files + test_abnormal_image_files
# In kết quả dự đoán
for i, image_file in enumerate(test_image_files):
    prediction = y_test[i]
    label = 'Class 1' if prediction >= 0.5 else 'Class 0'
    print(f"Image: {image_file}, Prediction: {prediction}, Label: {label}")
```
